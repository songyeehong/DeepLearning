## 필로우 기본 조작

```python
# 라이브러리 가져 오기
from PIL import Image
import numpy as np
```

```python
# 이미지 파일 가져 오기
image = Image.open('./sample_images/cat1.png', 'r')

# 이미지 파일 출력
image
```

```python
# 이미지 파일 크기 출력
print(image.size)
```

```python
# 이미지 파일의 픽셀값 추출
image_px = np.array(image)

# 이미지 파일의 픽셀값 확인
print(image_px[0][0])        # 왼쪽 맨위
print(image_px[184][272])    # 오른쪽 맨아래
```

```python
# 이미지 파일의 픽셀값 추출
image_px = np.array(image)

# 이미지 파일의 픽셀값 출력
print(image_px)
```

```python
# 이미지 픽셀값(배열)의 크기 출력
print(len(image_px))
```

```python
# 이미지를 1차원 배열로 변환하기
image_flatten = image_px.flatten().astype(np.float32) / 255.0
print(image_flatten)
```

```python
# 이미지 픽셀값(배열)의 크기 출력
print(len(image_flatten))
```

```python
# 이미지를 그레이스케일로 변환하기
gray_image = image.convert('L')

# 변환된 이미지 파일 출력
gray_image
```

```python
# 이미지 파일의 픽셀값 추출
gray_image_px = np.array(gray_image)

# 이미지 파일의 픽셀값 확인
print(gray_image_px[0][0])        # 왼쪽 맨위
print(gray_image_px[184][272])    # 오른쪽 맨아래
```

```python
# 이미지 파일 크기 출력
print(gray_image_px.size)
```

```python
# 이미지 파일의 픽셀값 출력
print(gray_image_px)
```

```python
# 이미지를 1차원 배열로 변환하기
gray_image_flatten = gray_image_px.flatten().astype(np.float32) / 255.0
print(gray_image_flatten)

# 이미지 픽셀값(배열)의 크기 출력
print(len(gray_image_flatten))
```

## 이미지 가공하기

```python
# 라이브러리 가져 오기
from PIL import ImageEnhance
```

```python
# 이미지의 채도 조정하기
conv1 = ImageEnhance.Color(image)
conv1_image = conv1.enhance(0.5)
conv1_image
```

```python
# 이미지의 명도 조정하기
conv2 = ImageEnhance.Brightness(image)
conv2_image = conv2.enhance(0.5)
conv2_image
```

```python
# 이미지의 콘트라스트 조정하기
conv3 = ImageEnhance.Contrast(image)
conv3_image = conv3.enhance(0.5)
conv3_image
```

```python
# 이미지의 날카로움 조정하기
conv4 = ImageEnhance.Sharpness(image)
conv4_image = conv4.enhance(2.0)
conv4_image
```

```python
# 이미지의 크기 조정하기
conv5_image = image.resize((128,128))
conv5_image
```

```python
# 이미지의 크기 출력
print(conv5_image.size)
```

```python
# 이미지의 방향 조정하기
#conv6_image = image.transpose(Image.FLIP_LEFT_RIGHT)
#conv6_image = image.transpose(Image.FLIP_TOP_BOTTOM)
#conv6_image = image.transpose(Image.ROTATE_90)
#conv6_image = image.transpose(Image.ROTATE_180)
conv6_image = image.transpose(Image.ROTATE_270)
conv6_image
```

```python
# 가공한 이미지 저장하기
conv6_image.save('./sample_images/new/conv6_5.png')
```
